```python
###Ques 1 :Login ans Logout
driver.get('https://www.instagram.com/')
driver.maximize_window()
a = driver.find_element_by_link_text('Log in')
a.click()
driver.implicitly_wait(10)
driver.refresh()
username = "SAMPLEUSERNAME"
password= "SAMPLEPASSWORD"
def login(usern,passw):
    username = driver.find_element_by_name('username')
    password = driver.find_element_by_name('password')
    username.send_keys(usern)
    password.send_keys(passw)
    driver.find_element_by_tag_name('button').submit()
    try:
        driver.find_element_by_xpath('//div[@class = "piCib"]/div[@class="mt3GC"]/button[last()]').click()
    except NoSuchElementException:
        return
        

def logout():
    driver.find_element_by_xpath('//div[contains(@class,"MWDvN")]//div[@class= "XrOey"][last()]/a').click()
    driver.find_element_by_xpath('//div[@class="AFWDX"]').click()
    driver.find_element_by_xpath('//div[contains(@class,pbNvD)]//button[contains(@class,aOOlW)][last()-1]').click()

def search(tosearch):
    driver.get('https://www.instagram.com/')
    driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/div').click()
    driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input').send_keys(tosearch)

login(username,password)
##logout()
```

```python
###Ques2
def allnames(tosearch):
    search(tosearch)
    ans = []
    for i in  driver.find_elements_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]'):
        data = BeautifulSoup(i.get_attribute('innerHTML'),'html.parser')
        if(data.div.div['class'][0] == 'RR-M-'):
            ans.append(data.find_all('div')[0].find_all('div')[1].span.string)
    return ans

NamesForFood = allnames("food")
for i in NamesForFood:
    print(i)
```

```python
###Ques4 Following and Unfollowing a handle
def followOrUnfollow(tosearch,option):
    driver.get('https://www.instagram.com/')
    search(tosearch)
    i = driver.find_element_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]')
    i.click()
    if(option):
        j = driver.find_element_by_xpath('//section//span[contains(@class,"vBF20")]/button')
        data = BeautifulSoup(j.get_attribute('innerHTML'),'html.parser')
        if(data.string == 'Follow'):
            j.click()
            print("Followed")
        else:
            print("Already Followed")
    else:
        j = driver.find_element_by_xpath('//section//span[contains(@class,"vBF20")]/button')
        data = BeautifulSoup(j.get_attribute('innerHTML'),'html.parser')
        if(data.string == "Following"):
            j.click()
            driver.find_element_by_xpath('//div[@class = "piCib"]/div[last()]/button').click()
            print("Unfollowed")
        else:
            print("Already Unfollowed")

followOrUnfollow('sodelhi',1)
followOrUnfollow('sodelhi',0)   ## 0 TO UNFOLLOW AND 1 TO FOLLOW
```

```python
###Ques 5 Liking top x(30) posts
def PostsWork(tosearch,x,option):
    search(tosearch)
    i = driver.find_element_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]')
    i.click()
    totalH = (max(x,24)-24)//3+1
    prevH = 0
    while(totalH):
        temp = driver.execute_script('return document.body.scrollHeight')
        if(temp == prevH):      ##To handle the case when x > no of posts
            break
        else:
            prevH = temp
        driver.execute_script('window.scrollTo(0,arguments[0])',temp)
        time.sleep(3)
        totalH-=1
        
    xstr = str(x)
    driver.execute_script('window.scrollTo(0,0)')
    
    li = driver.find_elements_by_xpath('//div[contains(@class,"_2z6nI")]//div[contains(@class,"v1Nh3")][position()<=' + xstr + ']')
    count = 0
    while(count < min(x,len(li))):
        driver.execute_script('window.scrollBy(0,30)')
        li[count].click()
        count+=1
        time.sleep(2)
        temp = driver.find_element_by_xpath('//div[@class ="eo2As "]/section/span/button')
        data = BeautifulSoup(temp.get_attribute('innerHTML'),'html.parser')
        if(option):
            print("Post", count , end = " ")
            if(data.svg.get('aria-label') == "Like"):
                driver.find_element_by_xpath('//div[@class ="eo2As "]/section/span/button').click()
                print("Liked")
            else:
                print("Post Already Liked")
        else:
            print("Post", count , end = " ")
            if(data.svg.get('aria-label') == "Unlike"):
                driver.find_element_by_xpath('//div[@class ="eo2As "]/section/span/button').click()
                print("Unliked")
            else:
                print("Post Already Unliked")
                
        time.sleep(2)
        driver.find_element_by_xpath('//button[@class = "ckWGn"]').click()

PostsWork('dilsefoodie',30,1)  ## 1 TO LIKE POSTS
PostsWork('dilsefoodie',30,0)  ## 0 TO UNLIKE POSTS
```

```python
###Ques6 Get Followers
def getFollowers(tosearch,countT):
    driver.get('https://www.instagram.com/')
    search(tosearch)
    i = driver.find_element_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]')
    i.click()
    time.sleep(3)
    driver.find_element_by_xpath('//ul[@class = "k9GMp "]/li[position() = 2]/a').click()
    time.sleep(3)
    if(countT == 0):              ##To get ALL the handles of FOLLOWERS
        countT = int(driver.find_element_by_xpath('//section[@class = "zwlfE"]/ul/li[position() = last()-1]/a/span').text)
    
    x = (countT)//12 + 1
    while(x != 1):
        x-=1
        temp = driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight')
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,arguments[0])',temp)
        time.sleep(1)        
        
    driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,0)')
    li = driver.find_elements_by_xpath('//div[@class = "PZuss"]/li[position() <=last()]')
    for i in range(min(countT, len(li))):
        data = BeautifulSoup(li[i].get_attribute('innerHTML'),'html.parser')
        li[i] = data.div.div.find_all(class_ = "enpQJ")[0].div.a.string        
    return li[:countT]

def getFollowing(tosearch,countT):
    driver.get('https://www.instagram.com/')
    search(tosearch)
    i = driver.find_element_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]')
    i.click()
    time.sleep(3)
    driver.find_element_by_xpath('//ul[@class = "k9GMp "]/li[position() = 3]/a').click()
    time.sleep(2)
    if(countT == 0):        ##TO GET ALL THE HANDLES USER IS FOLLOWING WE WILL SEND 0
        countT = int(driver.find_element_by_xpath('//section[@class = "zwlfE"]/ul/li[position() = last()]/a/span').text)
        
    x = (countT)//12 + 1    ##EACH SCROLL GIVES US 12 NAMES
    while(x != 1):
        x-=1
        temp = driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight')
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,arguments[0])',temp)
        time.sleep(1)        
        
    driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,0)')
    li = driver.find_elements_by_xpath('//div[@class = "PZuss"]/li[position() <=last()]')
    for i in range(min(countT, len(li))):
        data = BeautifulSoup(li[i].get_attribute('innerHTML'),'html.parser')
        li[i] = data.div.div.find_all(class_ = "enpQJ")[0].div.a.string
    return li[:countT]


def mutualFollowers(profile1,profilemine):
    driver.get('https://www.instagram.com/')
    search(profile1)
    i = driver.find_element_by_xpath('//div[@class = "drKGC"]/div[@class="fuqBx"]/a[contains(@class,"yCE8d")]')
    i.click()
    time.sleep(3)
    try :
        driver.find_element_by_xpath('//div[@class = "-vDIg"]/a[@class = "_32eiM"]').click()
    except NoSuchElementException:
        print("\n\nNo Mutual Followers")
        return []
    
    time.sleep(3)
    prevH = 0
    while(1):
        temp = driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight')
        if(temp == prevH):
            break
        else:
            prevH = temp
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,arguments[0])',temp)
        time.sleep(1)
            
    li = driver.find_elements_by_xpath('//div[@class = "PZuss"]//li[position() <=last()]')
    ans = []
    for i in li:
        data = BeautifulSoup(i.get_attribute('innerHTML'),'html.parser')
        if(data.div.find_all(class_="Igw0E rBNOH YBx95 ybXk5 _4EzTm soMvl")[0].button.string == "Following"):
            ans.append(data.div.find_all(class_="Igw0E IwRSH YBx95 vwCYk")[0].div.div.div.a.string)
    return ans
    

##Part1:
followersOfDilsefoodie = getFollowers('dilsefoodie',500)
followersOfFoodTalkIndia = getFollowers('foodtalkindia',500)

print("Followers of 'dilsefoodie':")
for i in range(len(followersOfDilsefoodie)):
    print("Follower",i+1,":",followersOfDilsefoodie[i])
    
print("\n\nFollowers of 'foodtalkindia':")
for i in range(len(followersOfFoodTalkIndia)):
    print("Follower",i+1,":",followersOfFoodTalkIndia[i])
    
    


##part2:
MyFollowers = getFollowers(username,0)       ##Sending 0 means to get all 
mutualFollowers = mutualFollowers('foodtalkindia',username)
ans = []
for i in mutualFollowers:
    if(i not in MyFollowers):
        ans.append(i)
if(len(ans)!= 0):
    print("\n\nFollowers of foodtalkindia who I am Following , and the are not following me: ")
    for i in ans:
        print(i)
```
